```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import csv
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# run a test city to explore the format of json response  
url = "http://api.openweathermap.org/data/2.5/weather?"
query_url = url + "appid=" + weather_api_key + "&q=" + "Melbourne" + "&units=imperial"
test_response = requests.get(query_url).json()   
test_response

```

```python
# create empty lists to store response climate data, then add these lists to dataframe  
city_list = list()
temp_list = list()
max_temp_list = list()
humidity_list = list()
cloud_list = list()
wind_list = list()
lat_list = list()
long_list = list()

```

```python
city_count = 0

# print log file to the terminal 
print("Beginning Data Retrieval ")
print("-----------------------------")

# use for loop to access climate information from the city list
for city in cities:
    try:
        query_url = url + "appid=" + weather_api_key + "&q=" + city + "&units=imperial"
        response = requests.get(query_url).json()   
        city_list.append(response["name"]) 
        temp_list.append(response["main"]["temp"])
        max_temp_list.append(response["main"]["temp_max"])
        humidity_list.append(response["main"]["humidity"]) 
        cloud_list.append(response["clouds"]["all"]) 
        wind_list.append(response["wind"]["speed"]) 
        lat_list.append(response["coord"]["lat"]) 
        long_list.append(response["coord"]["lon"])
    
        city_count += 1 
        print(f"Processing city {city}, total city number {city_count}") 
        
    # when there is no climatic data available for the city, skip it to avoid index/key error: 
    except(KeyError, IndexError):
        print("Skipped nowhere land...")
              
print("-----------------------------")
print(f"Data retrieval complete, total city number {city_count}")
print("-----------------------------")
    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
print(len(city_list))
print(len(long_list))
```

```python
# add list to the dataframe 
city_weather_list = pd.DataFrame({
                "City" : city_list,
                "Temp" : temp_list,
                "Max_temp": max_temp_list,
                "Cloudiness" : cloud_list,
                "Humidity" : humidity_list,
                "Wind Speed" : wind_list,                                 
                "Lat" : lat_list,
                "Long" : long_list                  
})

city_weather_list.reset_index()
```

```python
# export dataframe to csv
city_weather_list.to_csv("../output_data/city_weather_list.csv", index = False)

# remove null values so that all arrays have the same length

print("csv exported!")
city_weather_list.head(25)
#city_weather_list["Humidity"].max()
```

## Linear Regression

```python
# OPTIONAL: Create a function to create Linear Regression plots
def regression (value, lat):
    (slope, intercept, rvalue, pvalue, stderr) = linregress(value, lat)
    regression = value * slope + intercept
    line = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    plt.scatter(value, lat)
    plt.scatter(value, lat)
    plt.plot(value,regression,"r-")
    return line

```
